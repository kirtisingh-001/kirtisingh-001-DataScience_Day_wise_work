LAB 1: Python Fundamentals + File Handling
🎯 Objective:
Build a Student Result Processing System

🧩 Tasks:
Take input for 5 students (name + marks)

Calculate percentage & grade

Store results in a .txt file

Read file and display toppers

📤 Expected Output:
Structured file output

Sorted results

```python
students = []

for i in range(5):
    name = input("Enter name: ")
    marks = int(input("Enter marks: "))

    percent = marks
    if percent >= 90:
        grade = "A"
    elif percent >= 75:
        grade = "B"
    else:
        grade = "C"

    students.append((name, percent, grade))

# Save to file
with open("results.txt", "w") as f:
    for s in students:
        f.write(f"{s[0]} {s[1]} {s[2]}\n")

# Read and find topper
with open("results.txt", "r") as f:
    data = f.readlines()

topper = max(students, key=lambda x: x[1])
print("Topper:", topper)
```

LAB 2: Data Types + Control Flow
🎯 Objective:
Build a Data Analyzer Tool

🧩 Tasks:
Input a mixed list

Separate integers, floats, strings

Count frequency of each type

Print summary using loops

```python
data = [1, 2.5, "hi", 3, "hello", 4.5]

ints, floats, strings = [], [], []

for item in data:
    if isinstance(item, int):
        ints.append(item)
    elif isinstance(item, float):
        floats.append(item)
    else:
        strings.append(item)

print("Integers:", len(ints))
print("Floats:", len(floats))
print("Strings:", len(strings))
```

LAB 3: Functions + Recursion
🎯 Objective:
Create a Math Utility Module

🧩 Tasks:
Factorial using recursion

Fibonacci using recursion

Lambda for square & cube

Compare execution time

```python
# Factorial
def fact(n):
    if n == 1:
        return 1
    return n * fact(n-1)

# Fibonacci
def fib(n):
    if n <= 1:
        return n
    return fib(n-1) + fib(n-2)

# Lambda
square = lambda x: x*x
cube = lambda x: x*x*x

print(fact(5), fib(6), square(3), cube(3))
```

LAB 6: OOP + Inheritance
🎯 Objective:
Design an Employee Management System

🧩 Tasks:
Base class: Employee

Derived: Manager, Developer

Override salary method

Display employee details

```python
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

    def show(self):
        print(self.name, self.salary)

class Manager(Employee):
    def show(self):
        print("Manager:", self.name, self.salary)

e = Manager("John", 50000)
e.show()
```

LAB 7: Decorators + Generators

Optimize performance using advanced Python

🧩 Tasks:
Create decorator for logging

Create generator for large data

Compare memory usage

```python
def log(func):
    def wrapper():
        print("Running function")
        func()
    return wrapper

@log
def hello():
    print("Hello")

hello()

# Generator
def gen():
    for i in range(5):
        yield i

for i in gen():
    print(i)
```

LAB 10: Mini Project (FINAL)
🎯 Objective:
Build a Student Management System (End-to-End)

🧩 Features:
Add student

Search student

Store in CSV

Display analytics (topper, avg marks)




```python
from google.colab import files
uploaded = files.upload()
```

```python
import csv

FILE_NAME = "students.csv"

# Add Student
def add_student():
    name = input("Enter Name: ")
    marks = int(input("Enter Marks: "))

    with open(FILE_NAME, "a", newline="") as f:
        writer = csv.writer(f)
        writer.writerow([name, marks])

    print(" Student Added\n")

# Display All Students
def display_students():
    try:
        with open(FILE_NAME, "r") as f:
            reader = csv.reader(f)
            print("\n--- Student Records ---")
            for row in reader:
                print(row)
    except FileNotFoundError:
        print(" File not found!")

# Search Student
def search_student():
    name = input("Enter name to search: ")
    found = False

    with open(FILE_NAME, "r") as f:
        reader = csv.reader(f)
        for row in reader:
            if row[0].lower() == name.lower():
                print(" Found:", row)
                found = True

    if not found:
        print(" Student not found")

#  Analytics (Topper + Average)
def analytics():
    students = []

    with open(FILE_NAME, "r") as f:
        reader = csv.reader(f)
        for row in reader:
            try:
                students.append((row[0], int(row[1])))
            except:
                continue

    if not students:
        print("No data found")
        return

    topper = max(students, key=lambda x: x[1])
    avg = sum(s[1] for s in students) / len(students)

    print("\n Topper:", topper)
    print("Average Marks:", avg)

#  Menu
while True:
    print("\n1. Add Student")
    print("2. Display Students")
    print("3. Search Student")
    print("4. Analytics")
    print("5. Exit")

    choice = input("Enter choice: ")

    if choice == "1":
        add_student()
    elif choice == "2":
        display_students()
    elif choice == "3":
        search_student()
    elif choice == "4":
        analytics()
    elif choice == "5":
        print("Exiting...")
        break
    else:
        print("Invalid choice!")
```
