```python
# 🎯 Example 1: Sequential Multi-Agent System (Pipeline)
# 🧠 Scenario

# “A travel company has different employees (agents):

# Planner → decides steps

# Flight Agent → finds flights

# Weather Agent → checks weather

# Decision Agent → gives final answer”

# ================================
# AGENT 1: PLANNER
# ================================
def planner_agent(user_query):
    print("\n[Planner Agent] Creating plan...")
    return ["flight", "weather", "decision"]


# ================================
# AGENT 2: FLIGHT AGENT
# ================================
def flight_agent():
    print("\n[Flight Agent] Fetching flights...")
    return [
        {"airline": "IndiGo", "price": 4500},
        {"airline": "Air India", "price": 5200}
    ]


# ================================
# AGENT 3: WEATHER AGENT
# ================================
def weather_agent():
    print("\n[Weather Agent] Checking weather...")
    return {"condition": "Clear", "temp": 28}


# ================================
# AGENT 4: DECISION AGENT
# ================================
def decision_agent(flights, weather):
    print("\n[Decision Agent] Making decision...")

    cheapest = min(flights, key=lambda x: x["price"])

    if weather["condition"] == "Rain":
        return "Avoid travel due to bad weather"

    return f"Book {cheapest['airline']} at ₹{cheapest['price']}"


# ================================
# MAIN MULTI-AGENT SYSTEM
# ================================
def travel_multi_agent(user_query):
    print("User Query:", user_query)

    plan = planner_agent(user_query)

    flights = None
    weather = None

    for step in plan:
        if step == "flight":
            flights = flight_agent()

        elif step == "weather":
            weather = weather_agent()

        elif step == "decision":
            result = decision_agent(flights, weather)

    return result


# RUN
response = travel_multi_agent("Plan my trip Delhi to Mumbai")
print("\nFinal Answer:", response)
```

```python
# Example 2: Manager–Worker Multi-Agent System
# 🧠 Scenario

# “Now instead of fixed flow, we introduce a Manager Agent
# that assigns tasks dynamically to worker

# ================================
# WORKER AGENTS
# ================================
def flight_agent():
    print("[Flight Agent] Working...")
    return [{"airline": "IndiGo", "price": 4500},
            {"airline": "Air India", "price": 5200}]


def weather_agent():
    print("[Weather Agent] Working...")
    return {"condition": "Clear", "temp": 28}


# ================================
# MANAGER AGENT
# ================================
def manager_agent(user_query):
    print("\n[Manager Agent] Analyzing task...")

    tasks = []

    if "trip" in user_query.lower():
        tasks = ["flight", "weather"]

    return tasks


# ================================
# EXECUTION
# ================================
def run_system(user_query):
    print("User Query:", user_query)

    tasks = manager_agent(user_query)

    results = {}

    for task in tasks:
        if task == "flight":
            results["flights"] = flight_agent()

        elif task == "weather":
            results["weather"] = weather_agent()

    # Final decision
    cheapest = min(results["flights"], key=lambda x: x["price"])

    return f"Manager Decision: Book {cheapest['airline']} at ₹{cheapest['price']}"


# RUN
response = run_system("Plan my trip")
print("\nFinal Answer:", response)
```

```python
# ============================================
# 🏥 SCENARIO: Hospital Multi-Agent System
# ============================================

# “A hospital uses different employees (agents)
# to handle patient care in sequence.”

# --------------------------------------------
# AGENT 1: Intake Agent (Planner)
# --------------------------------------------
# - Collects patient symptoms and history
# - Decides next steps (diagnostic, treatment, decision)

# --------------------------------------------
# AGENT 2: Diagnostic Agent
# --------------------------------------------
# - Analyzes symptoms
# - Suggests possible diseases
# - Recommends tests (blood test, scan, etc.)

# --------------------------------------------
# AGENT 3: Treatment Agent
# --------------------------------------------
# - Suggests treatment options
#   (medication, therapy, surgery)
# - Considers patient condition and guidelines

# --------------------------------------------
# AGENT 4: Decision Agent
# --------------------------------------------
# - Combines all information
# - Gives final recommendation to patient


# ============================================
# FLOW OF SYSTEM
# ============================================

# Step 1:
# User → gives symptoms
# Example: "I have fever and cough"

# Step 2:
# Intake Agent → extracts symptoms + creates plan

# Step 3:
# Diagnostic Agent → finds possible illness

# Step 4:
# Treatment Agent → suggests treatment

# Step 5:
# Decision Agent → final advice


# ============================================
# FINAL OUTPUT
# ============================================

# Example:
# Diagnosis: Viral Infection
# Treatment: Paracetamol + Rest
# Advice: Stay hydrated and consult doctor if needed
from huggingface_hub import InferenceClient

client = InferenceClient(
    provider="auto",
    api_key=os.getenv("abcde")
)

# ================================
# HELPER FUNCTION
# ================================
def call_llm(prompt):
    try:
        response = client.chat.completions.create(
            model="meta-llama/Meta-Llama-3-8B-Instruct",
            messages=[
                {"role": "user", "content": prompt}
            ],
            max_tokens=200
        )
        return response.choices[0].message.content

    except Exception as e:
        return f"Error: {str(e)}"


# ================================
# WORKER AGENTS (AI BASED)
# ================================
def flight_agent(user_query):
    print("[Flight Agent] Working...")

    prompt = f"""
    User wants to travel: {user_query}

    Suggest 2 flight options with price (in INR).
    Format:
    Airline - Price
    """

    return call_llm(prompt)


def weather_agent(user_query):
    print("[Weather Agent] Working...")

    prompt = f"""
    Check weather for this travel plan:
    {user_query}

    Give condition and temperature.
    """

    return call_llm(prompt)


# ================================
# MANAGER AGENT (AI DECISION MAKER)
# ================================
def manager_agent(user_query):
    print("\n[Manager Agent] Analyzing task...")

    prompt = f"""
    User query: {user_query}

    Decide which tasks are needed from:
    - flight
    - weather

    Return ONLY list like:
    flight, weather
    """

    response = call_llm(prompt)

    # simple parsing
    tasks = []
    if "flight" in response.lower():
        tasks.append("flight")
    if "weather" in response.lower():
        tasks.append("weather")

    return tasks


# ================================
# FINAL DECISION AGENT
# ================================
def decision_agent(results):
    print("\n[Decision Agent] Making final decision...")

    prompt = f"""
    Based on results:
    {results}

    Give final recommendation for user.
    """

    return call_llm(prompt)


# ================================
# EXECUTION SYSTEM
# ================================
def run_system(user_query):
    print("User Query:", user_query)

    tasks = manager_agent(user_query)

    results = {}

    for task in tasks:
        if task == "flight":
            results["flights"] = flight_agent(user_query)

        elif task == "weather":
            results["weather"] = weather_agent(user_query)

    final = decision_agent(results)

    return final


# ================================
# USER INPUT LOOP
# ================================
while True:
    user_input = input("\nEnter your query (or type 'exit'): ")

    if user_input.lower() == "exit":
        break

    response = run_system(user_input)

    print("\n✅ Final Answer:\n", response)
```

```python
# ============================================
# 🚨 SCENARIO: Crisis Management Multi-Agent System
# ============================================

# “A company faces a data breach and multiple departments
# (agents) respond simultaneously to handle the crisis.”

# --------------------------------------------
# AGENT 1: Crisis Coordinator (Broadcaster)
# --------------------------------------------
# - Detects crisis
# - Broadcasts message to all agents:
#   "Data breach detected in customer database.
#    Immediate response required."
# - Sends message to all agents at the same time

# --------------------------------------------
# AGENT 2: IT Security Agent
# --------------------------------------------
# - Receives broadcast
# - Responds with technical actions:
#   • Isolate affected servers
#   • Patch vulnerabilities
#   • Start forensic analysis

# --------------------------------------------
# AGENT 3: Communications Agent
# --------------------------------------------
# - Receives broadcast
# - Handles communication:
#   • Draft internal memo
#   • Prepare press release
#   • Notify stakeholders

# --------------------------------------------
# AGENT 4: Finance Agent
# --------------------------------------------
# - Receives broadcast
# - Handles financial planning:
#   • Estimate financial impact
#   • Allocate emergency funds
#   • Review insurance coverage

# --------------------------------------------
# AGENT 5: Legal Agent
# --------------------------------------------
# - Receives broadcast
# - Handles legal compliance:
#   • Assess regulatory obligations
#   • Prepare compliance reports
#   • Advise on liability risks

# --------------------------------------------
# AGENT 6: HR Agent
# --------------------------------------------
# - Receives broadcast
# - Manages employees:
#   • Brief employees about situation
#   • Guide handling of customer queries
#   • Maintain morale and support

# --------------------------------------------
# AGENT 7: Decision Agent (Coordinator)
# --------------------------------------------
# - Collects responses from all agents
# - Integrates everything into final plan
# - Provides unified crisis response


# ============================================
# FLOW OF SYSTEM
# ============================================

# Step 1:
# Crisis Coordinator → broadcasts message

# Step 2:
# All agents (IT, Communication, Finance, Legal, HR)
# receive message simultaneously

# Step 3:
# Each agent responds with its domain-specific action

# Step 4:
# Decision Agent → combines all responses

# Step 5:
# Final crisis response plan generated


# ============================================
# FINAL OUTPUT (Example)
# ============================================

# "Servers isolated, communications prepared,
# financial impact assessed, compliance secured,
# employees briefed."
import os
from huggingface_hub import InferenceClient

# 🔐 hidden API key
client = InferenceClient(
    provider="auto",
    api_key=os.getenv("abcde")
)

# ================================
# HELPER FUNCTION
# ================================
def call_llm(prompt):
    try:
        response = client.chat.completions.create(
            model="meta-llama/Meta-Llama-3-8B-Instruct",
            messages=[
                {"role": "user", "content": prompt}
            ],
            max_tokens=200
        )
        return response.choices[0].message.content

    except Exception as e:
        return f"Error: {str(e)}"


# ================================
# AGENT 1: CRISIS COORDINATOR
# ================================
def crisis_coordinator():
    print("\n[Crisis Coordinator] Broadcasting message...")
    return "Data breach detected in customer database. Immediate response required."


# ================================
# WORKER AGENTS
# ================================
def it_security_agent(message):
    print("[IT Security Agent] Responding...")

    prompt = f"""
    Crisis: {message}

    What should IT Security do?
    """

    return call_llm(prompt)


def communications_agent(message):
    print("[Communications Agent] Responding...")

    prompt = f"""
    Crisis: {message}

    What should Communications team do?
    """

    return call_llm(prompt)


def finance_agent(message):
    print("[Finance Agent] Responding...")

    prompt = f"""
    Crisis: {message}

    What should Finance team do?
    """

    return call_llm(prompt)


def legal_agent(message):
    print("[Legal Agent] Responding...")

    prompt = f"""
    Crisis: {message}

    What should Legal team do?
    """

    return call_llm(prompt)


def hr_agent(message):
    print("[HR Agent] Responding...")

    prompt = f"""
    Crisis: {message}

    What should HR team do?
    """

    return call_llm(prompt)


# ================================
# DECISION AGENT
# ================================
def decision_agent(responses):
    print("\n[Decision Agent] Creating final plan...")

    prompt = f"""
    Combine these responses into a clear crisis response plan:

    {responses}

    Give a final structured plan.
    """

    return call_llm(prompt)


# ================================
# MAIN SYSTEM (BROADCAST FLOW)
# ================================
def run_crisis_system():
    # Step 1: broadcast
    message = crisis_coordinator()

    # Step 2: all agents receive SAME message
    responses = {}

    responses["IT"] = it_security_agent(message)
    responses["Communications"] = communications_agent(message)
    responses["Finance"] = finance_agent(message)
    responses["Legal"] = legal_agent(message)
    responses["HR"] = hr_agent(message)

    # Step 3: decision
    final_plan = decision_agent(responses)

    return final_plan


# ================================
# RUN
# ================================
if __name__ == "__main__":
    result = run_crisis_system()

    print("\n🚨 FINAL CRISIS RESPONSE PLAN:\n")
    print(result)
```
