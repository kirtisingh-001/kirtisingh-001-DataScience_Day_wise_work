```python
#Scenario 1: Online Course Completion Prediction
#We want to classify whether a student will complete an online course (1) or drop out (0) based on two features:
#Videos Watched – number of course videos the student watched
#Time Spent on Platform – total minutes spent on the learning platform

import numpy as np

# Training data
X = np.array([
    [2,15],
    [3,20],
    [8,60],
    [9,75]
])
y = np.array([[0],[0],[1],[1]])

# Normalize data
X = X / np.max(X, axis=0) #Normalizes data-Large values can make training slow or unstable

# Initialize weights
np.random.seed(1)
w1 = np.random.rand(2,3) #Weights from input → hidden layer
w2 = np.random.rand(3,1) #Weights from hidden → output

# Sigmoid function-converts any number into a value between 0 and 1.
def sigmoid(x):
    return 1/(1+np.exp(-x))

# Derivative of sigmoid-slope(for learning)
def sigmoid_derivative(x):
    return x*(1-x)

# Training
for i in range(5000):#will train 5000 times
    # Forward propagation
    #sending input data through the neural network to get predictions.
    hidden_input = np.dot(X,w1)
    hidden_output = sigmoid(hidden_input)# apply sigmoid activation function To convert values into non-linear outputs between 0 and 1.
    final_input = np.dot(hidden_output,w2)
    output = sigmoid(final_input)

    # Error
    error = y - output

    # Backpropagation -how much the output layer contributed to the error.
    d_output = error * sigmoid_derivative(output)
    error_hidden = d_output.dot(w2.T) #send the error backward to the hidden layer.Each hidden neuron now receives its share of the error
    d_hidden = error_hidden * sigmoid_derivative(hidden_output) #Without backpropagation, the network cannot learn.

    # Update weights- updates the weights of the neural network
    w2 += hidden_output.T.dot(d_output)
    w1 += X.T.dot(d_hidden)
#transpose helps produce the correct weight update matrix.
print("Predictions:")
print(output)
```
