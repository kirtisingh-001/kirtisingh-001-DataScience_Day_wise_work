```python
import matplotlib.pyplot as plt
# 1. Prepare the Data
months = ["January", "February", "March", "April"]
attendance = [85, 90, 95, 88] # Average students present
# 2. Create the Plot
plt.figure(figsize=(8, 5))
plt.plot(months, attendance, marker='o', linestyle='-', color='green', linewidth=2)
# 3. Label the axes and add a title
plt.title("Monthly Student Attendance Trend", fontsize=14)
plt.xlabel("Month", fontsize=12)
plt.ylabel("Average Students Present", fontsize=12)
# Show the plot
plt.show()
# Find the month with the highest attendance
max_attendance = max(attendance)
best_month = months[attendance.index(max_attendance)]
print(f"Highest Attendance: {best_month} with {max_attendance} students.")
# Check for steady increase or ups and downs
is_steady = all(attendance[i] < attendance[i+1] for i in range(len(attendance)-1))
if is_steady:
    print("Trend: There was a steady increase in attendance.")
else:
    print("Trend: There were ups and downs (fluctuations) in attendance.")
# Possible explanation for the dip in April
print("\nPossible explanation for the April dip:")
print("The dip in April might be explained by spring break, exam preparation periods, or seasonal illness.")
```

```python
import matplotlib.pyplot as plt
import numpy as np
ad_spend=np.array([2000,4000,6000,8000,10000,12000,14000,16000,18000])
sales=np.array([7000,15000,21000,25000,32000,36000,42000,47000,52000])
plt.scatter(ad_spend,sales,color='blue',alpha=0.7)
z=np.polyfit(ad_spend,sales,1)
p=np.poly1d(z)
plt.plot(ad_spend,p(ad_spend),label='trend')
plt.xlabel('Ad Spend')
plt.ylabel('Sales')
plt.title('Ad Spend vs Sales')
plt.legend()
plt.show()
```

```python
import random
heads=0
total=1000
for i in range(total):
  result=random.randint(0,1)
  if result==1:
    heads=heads+1
tails=total-heads
print("heads:",heads)
print("tails:",tails)
print("P(heads):",heads/total)
```
