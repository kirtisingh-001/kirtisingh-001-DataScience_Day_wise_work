```python
import torch
import torch.nn as nn
import torch.nn.functional as F

# Define the CNN Model
class DigitCNN(nn.Module):
    """
    A Convolutional Neural Network for handwritten digit recognition (MNIST)

    Input  : 1 x 28 x 28 image (grayscale digit)
    Output : 10 values (probability scores for digits 0-9)

    Architecture Overview:
    ----------------------
    Conv -> BatchNorm -> ReLU -> Pool
    Conv -> BatchNorm -> ReLU -> Pool
    Flatten
    Fully Connected
    Dropout
    Fully Connected -> Output
    """

    def __init__(self):
        # Call parent class constructor
        super(DigitCNN, self).__init__()

        # =========================================================
        # BLOCK 1
        # Convolution -> BatchNorm -> ReLU -> MaxPooling
        # =========================================================
        # Convolution Layer
        # Input channels = 1 (grayscale image)
        # Output channels = 32 (32 filters)
        # Kernel size = 3x3
        # Padding = 1 keeps spatial size same
        # Output size: 1 x 28 x 28 -> 32 x 28 x 28
        self.conv1 = nn.Conv2d(in_channels=1, out_channels=32, kernel_size=3, padding=1)
        self.bn1 = nn.BatchNorm2d(32)
        self.pool1 = nn.MaxPool2d(kernel_size=2, stride=2) # Becomes 32 x 14 x 14

        # BLOCK 2
        self.conv2 = nn.Conv2d(in_channels=32, out_channels=64, kernel_size=3, padding=1)
        self.bn2 = nn.BatchNorm2d(64)
        self.pool2 = nn.MaxPool2d(kernel_size=2, stride=2) # Becomes 64 x 7 x 7

        # Fully Connected Layers
        self.fc1 = nn.Linear(64 * 7 * 7, 128)
        self.dropout = nn.Dropout(0.5)
        self.fc2 = nn.Linear(128, 10)

    def forward(self, x):
        # Block 1
        x = self.pool1(F.relu(self.bn1(self.conv1(x))))

        # Block 2
        x = self.pool2(F.relu(self.bn2(self.conv2(x))))

        # Flatten
        x = x.view(-1, 64 * 7 * 7)

        # Fully Connected
        x = F.relu(self.fc1(x))
        x = self.dropout(x)
        x = self.fc2(x)

        return x

# Instantiate the model
model = DigitCNN()
print(model)
```
