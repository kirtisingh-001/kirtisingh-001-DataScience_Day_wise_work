```python
# 🎬 🎓 SCENARIO: “College Smart Assistant System”
# 🏫 Background Story

# A college builds an AI-powered student assistant.

# 👉 Students can ask:

# “What is my attendance?”
# “What are my marks?”

# 👉 Instead of manually checking portals,
# 👉 AI fetches it instantly.
# ================================
# STEP 1: Install Gradio
# ================================
!pip install gradio


# ================================
# STEP 2: Dummy Database (Tool Layer)
# ================================
students = {
    "101": {"name": "Rahul", "attendance": 85, "marks": 78},
    "102": {"name": "Priya", "attendance": 92, "marks": 88}
}


# ================================
# STEP 3: Tool Functions (MCP Tools)
# ================================
def get_attendance(student_id):
    if student_id in students:
        return f"📊 Attendance: {students[student_id]['attendance']}%"
    return "❌ Student not found"


def get_marks(student_id):
    if student_id in students:
        return f"📝 Marks: {students[student_id]['marks']}"
    return "❌ Student not found"


# ================================
# STEP 4: Security Layer
# ================================
def secure_access(user_id, requested_id):
    return user_id == requested_id


# ================================
# STEP 5: MCP Agent Logic
# ================================
def mcp_agent(message, student_id, history):

    # Simulate logged-in user (for demo)
    user_id = student_id

    # Security Check
    if not secure_access(user_id, student_id):
        return "🚫 Access Denied", history

    # Tool Invocation Logic
    message_lower = message.lower()

    if "attendance" in message_lower:
        response = get_attendance(student_id)

    elif "marks" in message_lower:
        response = get_marks(student_id)

    elif "hello" in message_lower or "hi" in message_lower:
        response = "👋 Hello! Ask me about attendance or marks."

    else:
        response = "🤖 I can help with attendance or marks."

    # Maintain chat history
    history.append((message, response))

    return "", history


# ================================
# STEP 6: Gradio Chat UI
# ================================
import gradio as gr

with gr.Blocks() as demo:

    gr.Markdown("# 🎓 Student MCP Agent (Gradio Version)")

    student_id = gr.Textbox(label="Enter Student ID (e.g., 101)")

    chatbot = gr.Chatbot()
    msg = gr.Textbox(label="Ask your question")

    state = gr.State([])

    msg.submit(
        mcp_agent,
        inputs=[msg, student_id, state],
        outputs=[msg, chatbot]
    )

# ================================
# STEP 7: Launch App (Public URL)
# ================================
demo.launch(share=True)
```

```python
# SCENARIO: “AI Banking Assistant with Role-Based Access”
# 🏦 Background Story

# A bank builds an AI assistant to help users:

# Check account balance
# View transactions
# Approve loans
# Manage customer accounts

# 👉 But not everyone can do everything

# 🧑‍🤝‍🧑 Roles in the Bank
# Role	Description
# 👤 Customer	Bank account holder
# 👨‍💼 Employee	Bank staff
# 🧑‍💼 Manager	Branch manager
# 🔐 Permissions (RBAC)
# Action	Customer	Employee	Manager
# View own balance	✅	✅	✅
# View others' accounts	❌	✅	✅
# Approve loan	❌	❌	✅
# View all transactions	❌	✅	✅


# ======================================
# STEP 1: Install Libraries
# ======================================
!pip install groq gradio


# ======================================
# STEP 2: Load API Key from Colab Secrets
# ======================================
from google.colab import userdata

groq_api_key = userdata.get("jklmps")

if not groq_api_key:
    raise ValueError(" GROQ_API_KEY not found in Colab Secrets")

from groq import Groq
client = Groq(api_key=groq_api_key)


# ======================================
# STEP 3: Dummy Bank Database
# ======================================
accounts = {
    "1001": {"name": "Amit", "balance": 50000},
    "1002": {"name": "Neha", "balance": 75000}
}


# ======================================
# STEP 4: Tool Functions (APIs)
# ======================================
def get_balance(account_id):
    if account_id in accounts:
        return f" Balance of {account_id}: ₹{accounts[account_id]['balance']}"
    return " Account not found"


def approve_loan(account_id):
    if account_id in accounts:
        return f"Loan approved for account {account_id}"
    return " Account not found"


# ======================================
# STEP 5: RBAC Security Layer
# ======================================
def secure_access(role, user_account, requested_account, action):

    # Manager → full access
    if role == "manager":
        return True

    # Employee → can view all but cannot approve loans
    elif role == "employee":
        if action == "approve_loan":
            return False
        return True

    # Customer → only own account, no loan approval
    elif role == "customer":
        return user_account == requested_account and action != "approve_loan"

    return False


# ======================================
# STEP 6: MCP Tool Decision via LLM
# ======================================
def decide_action(query):
    try:
        prompt = f"""
        You are an AI banking assistant.

        Decide which action to take:
        - get_balance
        - approve_loan

        Rules:
        - Balance related queries → get_balance
        - Loan approval queries → approve_loan

        Return ONLY the action name.

        Query: {query}
        """

        response = client.chat.completions.create(
            model="llama-3.3-70b-versatile",
            messages=[{"role": "user", "content": prompt}]
        )

        action = response.choices[0].message.content.strip().lower()
        return action

    except Exception as e:
        print(" Groq Error:", e)
        return "fallback"


# ======================================
# STEP 7: MCP Agent (Core Logic)
# ======================================
def banking_agent(message, role, user_account, requested_account, history):

    # Input validation
    if not user_account:
        response = " Please enter your account ID"
        history.append((message, response))
        return "", history

    if not requested_account:
        requested_account = user_account  # default to own account

    # Step 1: LLM decides action
    action = decide_action(message)

    # Step 2: RBAC Security Check
    if not secure_access(role, user_account, requested_account, action):
        response = " Access Denied: You are not authorized"

    else:
        # Step 3: Tool Invocation
        if "balance" in action:
            response = get_balance(requested_account)

        elif "loan" in action:
            response = approve_loan(requested_account)

        # Fallback if LLM fails
        elif action == "fallback":
            msg_lower = message.lower()
            if "balance" in msg_lower:
                response = get_balance(requested_account)
            elif "loan" in msg_lower:
                response = approve_loan(requested_account)
            else:
                response = " Could not understand request"

        else:
            response = " Try asking about balance or loan"

    # Save chat history
    history.append((message, response))

    return "", history


# ======================================
# STEP 8: Gradio UI
# ======================================
import gradio as gr

with gr.Blocks() as demo:

    gr.Markdown("# AI Banking Assistant (MCP + RBAC + Groq)")

    role = gr.Dropdown(
        ["customer", "employee", "manager"],
        label="Select Role"
    )

    user_account = gr.Textbox(label="Your Account ID (e.g., 1001)")
    requested_account = gr.Textbox(label="Target Account ID (optional)")

    chatbot = gr.Chatbot(height=400)
    msg = gr.Textbox(label="Ask your question")

    state = gr.State([])

    msg.submit(
        banking_agent,
        inputs=[msg, role, user_account, requested_account, state],
        outputs=[msg, chatbot]
    )


# ======================================
# STEP 9: Launch App
# ======================================
demo.launch(share=True)
```

```python
# ======================================
# UNIVERSITY SMART ASSISTANT (RBAC + MCP + LLM)
# ======================================

# ================================
# STEP 1: Install Libraries
# ================================
!pip install groq gradio


# ================================
# STEP 2: Load API Key
# ================================
from google.colab import userdata
from groq import Groq

api_key = userdata.get("jklmps")

if not api_key:
    raise ValueError("Groq API key not found")

client = Groq(api_key=api_key)


# ================================
# STEP 3: Dummy Database
# ================================
students = {
    "501": {"name": "Rahul", "attendance": 85, "results": "A"},
    "502": {"name": "Priya", "attendance": 92, "results": "A+"}
}


# ================================
# STEP 4: Tool Functions
# ================================
def get_attendance(student_id):
    if student_id in students:
        return f"Attendance: {students[student_id]['attendance']}%"
    return "Student not found"


def get_results(student_id):
    if student_id in students:
        return f"Exam Result: {students[student_id]['results']}"
    return "Student not found"


def update_schedule():
    return "Course schedule updated successfully"


def approve_registration(student_id):
    if student_id in students:
        return f"Registration approved for student {student_id}"
    return "Student not found"


# ================================
# STEP 5: RBAC Security
# ================================
def secure_access(role, user_id, requested_id, action):

    # Admin → full access
    if role == "admin":
        return True

    # Faculty → can view all + update schedule
    elif role == "faculty":
        if action == "approve_registration":
            return False
        return True

    # Student → only own data
    elif role == "student":
        return user_id == requested_id and action not in ["update_schedule", "approve_registration"]

    return False


# ================================
# STEP 6: LLM Decision
# ================================
def decide_action(query):

    prompt = f"""
    Decide action:
    get_attendance
    get_results
    update_schedule
    approve_registration

    Query: {query}

    Return only one action.
    """

    try:
        response = client.chat.completions.create(
            model="llama-3.3-70b-versatile",
            messages=[{"role": "user", "content": prompt}]
        )

        return response.choices[0].message.content.strip().lower()

    except:
        return "fallback"


# ================================
# STEP 7: Agent Logic
# ================================
def university_agent(message, role, user_id, requested_id, history):

    if not user_id:
        history.append((message, "Enter your ID"))
        return "", history

    if not requested_id:
        requested_id = user_id

    action = decide_action(message)

    if not secure_access(role, user_id, requested_id, action):
        response = "Access denied"

    else:
        if action == "get_attendance":
            response = get_attendance(requested_id)

        elif action == "get_results":
            response = get_results(requested_id)

        elif action == "update_schedule":
            response = update_schedule()

        elif action == "approve_registration":
            response = approve_registration(requested_id)

        elif action == "fallback":
            msg = message.lower()
            if "attendance" in msg:
                response = get_attendance(requested_id)
            elif "result" in msg:
                response = get_results(requested_id)
            elif "schedule" in msg:
                response = update_schedule()
            elif "approve" in msg:
                response = approve_registration(requested_id)
            else:
                response = "Could not understand request"

        else:
            response = "Ask about attendance, results, schedule or registration"

    history.append((message, response))
    return "", history


# ================================
# STEP 8: Gradio UI
# ================================
import gradio as gr

with gr.Blocks() as demo:

    gr.Markdown("University Smart Assistant")

    role = gr.Dropdown(
        ["student", "faculty", "admin"],
        label="Select Role"
    )

    user_id = gr.Textbox(label="Your ID (e.g. 501)")
    requested_id = gr.Textbox(label="Target Student ID (optional)")

    chatbot = gr.Chatbot()
    msg = gr.Textbox(label="Ask your question")

    state = gr.State([])

    msg.submit(
        university_agent,
        inputs=[msg, role, user_id, requested_id, state],
        outputs=[msg, chatbot]
    )


# ================================
# STEP 9: Launch
# ================================
demo.launch(share=True)
```

```python
# ======================================
# RETAIL SMART ASSISTANT (RBAC + MCP + LLM)
# ======================================

# ================================
# STEP 1: Install Libraries
# ================================
!pip install groq gradio


# ================================
# STEP 2: Load API Key
# ================================
from google.colab import userdata
from groq import Groq

api_key = userdata.get("jklmps")

if not api_key:
    raise ValueError("Groq API key not found")

client = Groq(api_key=api_key)


# ================================
# STEP 3: Dummy Data
# ================================
customers = {
    "C101": {"name": "Ravi", "history": ["Shirt", "Shoes", "Watch"]},
    "C102": {"name": "Anita", "history": ["Bag", "Dress"]}
}

inventory = {
    "shirt": 20,
    "shoes": 15,
    "watch": 10,
    "bag": 25
}


# ================================
# STEP 4: Tool Functions
# ================================
def get_purchase_history(customer_id):
    if customer_id in customers:
        items = ", ".join(customers[customer_id]["history"])
        return f"Purchase History: {items}"
    return "Customer not found"


def check_inventory(product):
    product = product.lower()
    if product in inventory:
        return f"Stock for {product}: {inventory[product]}"
    return "Product not found"


def approve_supplier_order():
    return "Supplier order approved"


def manage_schedule():
    return "Employee schedule updated"


# ================================
# STEP 5: RBAC Security
# ================================
def secure_access(role, user_id, requested_id, action):

    # Manager → full access
    if role == "manager":
        return True

    # Staff permissions
    elif role == "staff":
        if action in ["approve_supplier_order", "manage_schedule"]:
            return False
        return True

    # Customer permissions
    elif role == "customer":
        if action != "get_purchase_history":
            return False
        return user_id == requested_id

    return False


# ================================
# STEP 6: LLM Decision
# ================================
def decide_action(query):

    prompt = f"""
    Choose action:
    get_purchase_history
    check_inventory
    approve_supplier_order
    manage_schedule

    Query: {query}

    Return only action name.
    """

    try:
        response = client.chat.completions.create(
            model="llama-3.3-70b-versatile",
            messages=[{"role": "user", "content": prompt}]
        )

        return response.choices[0].message.content.strip().lower()

    except:
        return "fallback"


# ================================
# STEP 7: Extract product (simple)
# ================================
def extract_product(message):
    words = message.lower().split()
    for w in words:
        if w in inventory:
            return w
    return None


# ================================
# STEP 8: Agent Logic
# ================================
def retail_agent(message, role, user_id, requested_id, history):

    if not user_id:
        history.append((message, "Enter your ID"))
        return "", history

    if not requested_id:
        requested_id = user_id

    action = decide_action(message)

    if not secure_access(role, user_id, requested_id, action):
        response = "Access denied"

    else:
        if action == "get_purchase_history":
            response = get_purchase_history(requested_id)

        elif action == "check_inventory":
            product = extract_product(message)
            if product:
                response = check_inventory(product)
            else:
                response = "Specify product name"

        elif action == "approve_supplier_order":
            response = approve_supplier_order()

        elif action == "manage_schedule":
            response = manage_schedule()

        elif action == "fallback":
            msg = message.lower()
            if "history" in msg:
                response = get_purchase_history(requested_id)
            elif "inventory" in msg:
                product = extract_product(message)
                response = check_inventory(product) if product else "Specify product"
            elif "approve" in msg:
                response = approve_supplier_order()
            elif "schedule" in msg:
                response = manage_schedule()
            else:
                response = "Could not understand request"

        else:
            response = "Ask about purchase history, inventory, orders or schedules"

    history.append((message, response))
    return "", history


# ================================
# STEP 9: UI
# ================================
import gradio as gr

with gr.Blocks() as demo:

    gr.Markdown("Retail Smart Assistant")

    role = gr.Dropdown(
        ["customer", "staff", "manager"],
        label="Select Role"
    )

    user_id = gr.Textbox(label="Your ID (e.g. C101)")
    requested_id = gr.Textbox(label="Target Customer ID (optional)")

    chatbot = gr.Chatbot()
    msg = gr.Textbox(label="Ask your query")

    state = gr.State([])

    msg.submit(
        retail_agent,
        inputs=[msg, role, user_id, requested_id, state],
        outputs=[msg, chatbot]
    )


# ================================
# STEP 10: Launch
# ================================
demo.launch(share=True)
```

```python
# ======================================
# HEALTHCARE RESEARCH ASSISTANT SYSTEM
# (Parallel + MCP + LLM)
# ======================================

# ================================
# STEP 1: Install Libraries
# ================================
!pip install groq gradio


# ================================
# STEP 2: Setup API Key
# ================================
from google.colab import userdata
from groq import Groq

api_key = userdata.get("jklmps")

if not api_key:
    raise ValueError("Groq API key not found")

client = Groq(api_key=api_key)


# ================================
# STEP 3: Dummy Data Sources
# ================================
medical_news = {
    "diabetes": "New insulin therapies show improved glucose control.",
    "cancer": "Immunotherapy is advancing cancer treatment outcomes."
}

clinical_trials = {
    "diabetes": "Trial shows 30% improvement with new drug.",
    "cancer": "Phase 3 trial reports increased survival rates."
}

company_profiles = {
    "pfizer": "Pfizer is a leading pharmaceutical company focusing on vaccines.",
    "novartis": "Novartis specializes in innovative medicines."
}


# ================================
# STEP 4: Tool Functions
# ================================
def fetch_medical_news(topic):
    return medical_news.get(topic, "No news found")


def fetch_clinical_trials(topic):
    return clinical_trials.get(topic, "No trial data found")


def fetch_company_profile(company):
    return company_profiles.get(company.lower(), "No company data found")


# ================================
# STEP 5: Parallel Data Collection
# ================================
import concurrent.futures

def collect_data(query):

    query_lower = query.lower()

    topic = "diabetes" if "diabetes" in query_lower else "cancer"
    company = "pfizer" if "pfizer" in query_lower else "novartis"

    with concurrent.futures.ThreadPoolExecutor() as executor:
        news_future = executor.submit(fetch_medical_news, topic)
        trial_future = executor.submit(fetch_clinical_trials, topic)
        company_future = executor.submit(fetch_company_profile, company)

        news = news_future.result()
        trials = trial_future.result()
        company_info = company_future.result()

    return news, trials, company_info


# ================================
# STEP 6: LLM Analysis
# ================================
def analyze_data(query, news, trials, company):

    prompt = f"""
    You are a medical research assistant.

    Query: {query}

    Data:
    Medical News: {news}
    Clinical Trials: {trials}
    Company Info: {company}

    Generate a professional research summary.
    """

    response = client.chat.completions.create(
        model="llama-3.3-70b-versatile",
        messages=[{"role": "user", "content": prompt}]
    )

    return response.choices[0].message.content


# ================================
# STEP 7: Agent Logic
# ================================
def research_agent(message, history):

    # Step 1: Collect data in parallel
    news, trials, company = collect_data(message)

    # Step 2: LLM analysis
    report = analyze_data(message, news, trials, company)

    history.append((message, report))

    return "", history


# ================================
# STEP 8: Gradio UI
# ================================
import gradio as gr

with gr.Blocks() as demo:

    gr.Markdown("Healthcare Research Assistant")

    chatbot = gr.Chatbot()
    msg = gr.Textbox(label="Enter research query")

    state = gr.State([])

    msg.submit(
        research_agent,
        inputs=[msg, state],
        outputs=[msg, chatbot]
    )


# ================================
# STEP 9: Launch
# ================================
demo.launch(share=True)
```

```python
# ======================================
# AI IT HELPDESK ASSISTANT SYSTEM
# ======================================

# ================================
# STEP 1: Install Dependencies
# ================================
!pip install groq


# ================================
# STEP 2: Load API Key
# ================================
from google.colab import userdata
from groq import Groq
import asyncio
import random
import logging

logging.basicConfig(level=logging.INFO)

api_key = userdata.get("jklmps")

if not api_key:
    raise ValueError("Groq API key not found")

client = Groq(api_key=api_key)


# ================================
# STEP 3: Tool Functions (Async)
# ================================
async def page_security_team(payload):
    return "Security team notified"


async def create_ticket(payload):
    return f"Ticket created: {payload}"


async def check_network(payload):
    return random.choice(["Network degraded", "Network stable"])


async def alert_network_team(payload):
    return "Network team alerted"


async def get_user(payload):
    return "employee_001"


async def reset_access(payload):
    return "Access permissions reset"


async def query_main_db(payload):
    if random.random() < 0.7:
        raise Exception("Database failure")
    return "Data from main database"


async def query_backup_db(payload):
    return "Data from backup database"


# ================================
# STEP 4: LLM Classification
# ================================
def classify_issue(issue):

    prompt = f"""
    Classify issue into one category:
    network, hardware, software, security, access

    Issue: {issue}

    Return only category name.
    """

    response = client.chat.completions.create(
        model="llama-3.3-70b-versatile",
        messages=[{"role": "user", "content": prompt}]
    )

    return response.choices[0].message.content.strip().lower()


# ================================
# STEP 5: Routing Logic
# ================================
async def it_agent(issue, severity):

    category = classify_issue(issue)
    print("Category:", category)

    if category == "security":
        await page_security_team({"issue": issue})
        return await create_ticket({"type": "security", "priority": "high"})

    elif category == "network" and severity == "high":
        status = await check_network({})
        if "degraded" in status.lower():
            return await alert_network_team({"issue": issue})
        else:
            return await create_ticket({"type": "network", "priority": "high"})

    elif category == "access":
        user = await get_user({"query": issue})
        return await reset_access({"user": user})

    else:
        return await create_ticket({"type": "general", "priority": "medium"})


# ================================
# STEP 6: Retry Logic
# ================================
async def retry_call(primary, fallback=None, retries=3):

    delay = 1

    for i in range(retries):
        try:
            return await primary({})
        except Exception as e:
            logging.warning(f"Retry {i+1} failed: {e}")

            if i == retries - 1:
                if fallback:
                    logging.info("Using fallback")
                    return await fallback({})
                raise

            await asyncio.sleep(delay)
            delay *= 2


# ================================
# STEP 7: Demo Run
# ================================
async def run_demo():

    print("IT Helpdesk Demo")

    result = await it_agent(
        issue="VPN not working",
        severity="high"
    )

    print("Routing Result:", result)

    print("\nRetry Demo")

    data = await retry_call(
        query_main_db,
        fallback=query_backup_db
    )

    print("Data Result:", data)


# ================================
# STEP 8: Execute
# ================================
await run_demo()
```
