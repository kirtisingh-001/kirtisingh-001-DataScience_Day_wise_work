```python
import pandas as pd
import random

df = pd.read_csv("music_genre_dataset.csv")

data = df.values.tolist()

random.shuffle(data)

total = len(data)

train_end = int(0.7 * total)
val_end = int(0.85 * total)

train_data = data[:train_end]
val_data = data[train_end:val_end]
test_data = data[val_end:]

train_df = pd.DataFrame(train_data, columns=df.columns)
val_df = pd.DataFrame(val_data, columns=df.columns)
test_df = pd.DataFrame(test_data, columns=df.columns)

train_df.to_csv("train.csv", index=False)
val_df.to_csv("validation.csv", index=False)
test_df.to_csv("test.csv", index=False)

print("Training set size:", len(train_df))
print("Validation set size:", len(val_df))
print("Test set size:", len(test_df))
```

```python
import pandas as pd

df = pd.read_csv("Dataset  - Sheet1.csv")

exceptional = df[df["Score"] > 90]
sales_only = df[df["Department"] == "Sales"]

sorted_score = df.sort_values(by="Score", ascending=False)
sorted_dept_score = df.sort_values(by=["Department", "Score"], ascending=[True, False])

avg_dept = df.groupby("Department")["Score"].mean()
max_quarter = df.groupby("Quarter")["Score"].max()

print("Exceptional Performers:\n", exceptional)
print("\nSales Department:\n", sales_only)
print("\nSorted by Score:\n", sorted_score)
print("\nSorted by Department and Score:\n", sorted_dept_score)
print("\nAverage Score by Department:\n", avg_dept)
print("\nMaximum Score by Quarter:\n", max_quarter)
```
