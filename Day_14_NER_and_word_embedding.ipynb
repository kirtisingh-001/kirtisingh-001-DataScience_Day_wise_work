```python
# # 📘 Scenario: Corporate Expansion Analysis
# # Context:
# # You are part of a business intelligence team at a multinational company. Your manager has asked you to analyze press
# releases and news articles to identify key entities such as organizations, people, locations, dates, and numbers. This
#  information will help the company track competitors’ moves and plan strategy.
# # Task:
# # You receive the following press release excerpt:
# # "Apple Inc. plans a new office in Hyderabad, India. Tim Cook announced this in March 2023. The site will create 5,000 jobs
# and focus on innovative technologies in supply chain analytics."
# # Your job is to run Named Entity Recognition (NER) using spaCy to automatically extract structured insights.


import spacy
# Load the small English model
nlp = spacy.load("en_core_web_sm")
text = """
Apple Inc. plans a new office in Hyderabad, India. Tim Cook announced this in March 2023.
The site will create 5,000 jobs and focus on innovative technologies in supply chain analytics.
"""
# Load the spaCy NLP pipeline (make sure the model is installed)
# Example: python -m spacy download en_core_web_sm
import spacy
nlp = spacy.load("en_core_web_sm")
# Example text for entity extraction
text = """
Apple Inc. was founded by Steve Jobs and Steve Wozniak in California.
It is headquartered in Cupertino and has offices in New York and London.
In 2023, Apple reported revenue of $394 billion.
"""
# Process the text with spaCy NLP pipeline
# This performs tokenization, POS tagging, dependency parsing, and NER
doc = nlp(text)
# Print all named entities detected in the text
# Each entity will show:
# 1. The actual entity text
# 2. The entity label (type of entity)
print("Named Entities (text, label):")
for ent in doc.ents:
    # ent.text  -> the detected entity
    # ent.label_ -> the category of entity (ORG, PERSON, GPE etc.)
    print(f"{ent.text:25} -> {ent.label_}")
# Optional Section: Group entities by their label
# This helps summarize entities under each category
# Example:
# PERSON: Steve Jobs, Steve Wozniak
# ORG: Apple
# GPE: California, New York
from collections import defaultdict
# defaultdict automatically creates empty lists for new keys
by_label = defaultdict(list)
# Loop through entities and group them by label
for ent in doc.ents:
    by_label[ent.label_].append(ent.text)
# Print grouped entities
# sorted(set(items)) removes duplicates and sorts results
print("\nEntities grouped by label:")

for label, items in by_label.items():
    print(f"{label}: {sorted(set(items))}")
```

```python
import nltk
from gensim.models import Word2Vec
from nltk.tokenize import word_tokenize

# 1. Setup: Download the necessary tokenizer
nltk.download('punkt')
nltk.download('punkt_tab')

# 2. Manufacturing Corpus
# In a real scenario, this would be thousands of maintenance logs or telemetry reports.
corpus = [
    "Predictive maintenance reduces downtime and improves machine longevity.",
    "Real-time telemetry provides data on sensor performance and temperature.",
    "Quality control ensures that every product meets factory standards.",
    "Supply chain logistics are optimized using AI and machine learning.",
    "Sensors collect telemetry data to trigger maintenance alerts.",
    "The factory uses quality control and telemetry to monitor production."
]

# 3. Preprocessing: Tokenize and lowercase the text
tokenized_corpus = [word_tokenize(sentence.lower()) for sentence in corpus]

# 4. Train the Word2Vec Model
# vector_size: dimensions of the word vectors
# window: maximum distance between the current and predicted word
# min_count: ignores words with total frequency lower than this
model = Word2Vec(sentences=tokenized_corpus, vector_size=50, window=5, min_count=1, workers=4)

# 5. Analyze Word Similarities
def analyze_manufacturing_terms(term):
    print(f"\n--- Insights for '{term}' ---")
    try:
        # Finding the top 3 most similar words based on context
        similar_words = model.wv.most_similar(term, topn=3)
        for word, score in similar_words:
            print(f"Similarity with '{word}': {score:.4f}")
    except KeyError:
        print(f"Term '{term}' not found in the corpus.")

# Analyze key technical concepts
analyze_manufacturing_terms("telemetry")
analyze_manufacturing_terms("maintenance")
analyze_manufacturing_terms("quality")

# 6. Measuring Distance (Conceptual connection)
distance = model.wv.distance("telemetry", "data")
print(f"\nSemantic distance between 'telemetry' and 'data': {distance:.4f}")
```

```python
# Code Demo (Trigram counts)
from collections import defaultdict

corpus = "the cat sat on the mat the cat lay on the rug the dog barked loudly"
tokens = corpus.split()

# Build trigram counts
trigram_counts = defaultdict(lambda: defaultdict(int))
for i in range(len(tokens)-2):
    context = (tokens[i], tokens[i+1])
    next_word = tokens[i+2]
    trigram_counts[context][next_word] += 1

# Predict next word after "the cat"
context = ("the", "cat")
print("Next word predictions for context:", context)
for word, count in trigram_counts[context].items():
    print(f"{word} -> {count}")
```

```python
#📘 Scenario: Healthcare Research Insights
#Context:
#You are part of a data science team in a healthcare research institute. Your manager wants you to analyze medical reports and research papers to explore how different technical concepts (like diagnosis, treatment, symptoms, patients, medication) are related. By training a small Word2Vec model, you can uncover semantic similarities between words and measure how closely they are connected. This helps in building clinical decision support systems, medical knowledge graphs, or even automated literature reviews.
#Task:
#Given a short corpus describing healthcare operations, train a Word2Vec model and analyze word similarities.

import nltk
from gensim.models import Word2Vec
from nltk.tokenize import word_tokenize

# 1. Setup: Download the necessary tokenizer
nltk.download('punkt')
nltk.download('punkt_tab')

# 2. Healthcare Research Corpus
# This represents a small collection of medical report abstracts
corpus = [
    "Early diagnosis of symptoms improves patient recovery rates.",
    "The treatment protocol includes specific medication and physical therapy.",
    "Patients reporting chronic symptoms require a specialized diagnosis.",
    "New medication shows promising results in clinical treatment trials.",
    "Effective diagnosis is the first step toward a successful treatment plan.",
    "Medical researchers analyze patient data to identify common symptoms."
]

# 3. Preprocessing: Tokenize and lowercase the text
tokenized_corpus = [word_tokenize(sentence.lower()) for sentence in corpus]

# 4. Train the Word2Vec Model
# vector_size: 100 dimensions for medical nuances
# window: 5 (looking at 5 words before and after)
# min_count: 1 (include all words in this small sample)
model = Word2Vec(sentences=tokenized_corpus, vector_size=100, window=5, min_count=1, workers=4)

# 5. Analyze Semantic Similarities
def analyze_medical_terms(term):
    print(f"\n--- Clinical Insights for '{term}' ---")
    try:
        # Finding the top 3 most semantically related words
        similar_words = model.wv.most_similar(term, topn=3)
        for word, score in similar_words:
            print(f"Similarity with '{word}': {score:.4f}")
    except KeyError:
        print(f"Term '{term}' not found in the research corpus.")

# Analyze key clinical concepts
analyze_medical_terms("diagnosis")
analyze_medical_terms("treatment")
analyze_medical_terms("symptoms")

# 6. Measuring "Semantic Distance"
# Lower distance means the concepts are more closely linked in the literature
dist = model.wv.distance("medication", "treatment")
print(f"\nSemantic distance between 'medication' and 'treatment': {dist:.4f}")
```

```python
# ==========================================
# UNIVERSITY AI HELPDESK CHATBOT
# Demonstrates:
# Generative AI
# Large Language Models (LLMs)
# Tokenization
# Transformer Inference
# Prompt Engineering
# ==========================================

# STEP 1: Install required libraries
!pip install transformers torch

# STEP 2: Import libraries
from transformers import pipeline, GPT2Tokenizer

# STEP 3: Load pretrained LLM model (GPT-2)
generator = pipeline("text-generation", model="gpt2")

# STEP 4: Load tokenizer (for tokenization demonstration)
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")

# ------------------------------------------
# STEP 5: Define University Knowledge Base
# ------------------------------------------

knowledge_base = """
University Helpdesk Information

Hostel Admission Requirements:
- Admission confirmation letter
- Government ID proof
- Two passport size photographs

Course Registration:
- Registration happens online through the university portal
- Students must register before the semester deadline

Fee Payment:
- Fees can be paid through the student dashboard
- Online payment methods include debit card, credit card, and net banking

Examination Schedule:
- Semester exams usually start in December and May

Library Access:
- Students must carry their university ID card
"""

# ------------------------------------------
# STEP 6: Student Question
# ------------------------------------------

question = "How can I pay my university fees?"

# ------------------------------------------
# STEP 7: Prompt Engineering
# ------------------------------------------

prompt = f"""
You are a helpful AI assistant working for a university helpdesk.

Instructions:
- Answer clearly and politely.
- Use the information from the knowledge base.
- If the answer is not available, say "Please contact the university helpdesk".

Knowledge Base:
{knowledge_base}

Student Question:
{question}

Answer:
"""

print("PROMPT SENT TO MODEL:")
print(prompt)

# ------------------------------------------
# STEP 8: Tokenization Demonstration
# ------------------------------------------

tokens = tokenizer.tokenize(prompt)

print("\nTOKENIZATION OUTPUT:")
print("Total Tokens:", len(tokens))
print(tokens[:20])  # show first 20 tokens

# ------------------------------------------
# STEP 9: Transformer Inference
# ------------------------------------------

response = generator(
    prompt,
    max_length=200,
    temperature=0.7,
    num_return_sequences=1
)

# ------------------------------------------
# STEP 10: Generated Answer
# ------------------------------------------

print("\nAI GENERATED RESPONSE:")
print(response[0]["generated_text"])
```
